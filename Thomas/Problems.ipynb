### a)

```python
def downsample(x, D, offset):
    return x[offset::D]

n = np.arange(0,51)
x_n = np.sin(0.2*np.pi*n)

D = 4
k0 = 0
k2 = 2

x_d_k0 = downsample(x_n,D,k0)
x_d_k2 = downsample(x_n,D,k2)

# Original Signal
plt.figure()
plt.subplot(3,1,1)
plt.stem(n,x_n)
plt.title('Original Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

# Downsampled Signal with k = 0
plt.subplot(3, 1, 2)
plt.stem(n[k0::D], x_d_k0)
plt.title('Downsampled Signal with k = 0')
plt.xlabel('n')
plt.ylabel('x[n]')

# Downsampled Signal with k = 2
plt.subplot(3, 1, 3)
plt.stem(n[k2::D], x_d_k2)
plt.title('Downsampled Signal with k = 2')
plt.xlabel('n')
plt.ylabel('x[n]')

plt.tight_layout()
plt.show()
```

## 12.19

```python
from scipy.fft import fft, fftshift
# Step 1: Generate the signal x[n] = 0.9^n * u[n]
n = np.arange(0, 100)  # n from 0 to 99 for better resolution in FFT
x_n = 0.9**n

# Step 2: Compute the spectrum of x[n] using FFT
X_f = fftshift(fft(x_n))
X_magnitude = np.abs(X_f)

# Step 3: Downsample the signal by a factor of D = 3
D = 3
x_D_n = x_n[::D]

# Step 4: Compute the spectrum of the downsampled signal x_D[n]
X_D_f = fftshift(fft(x_D_n, n=len(x_n)))  # Use the same length for FFT comparison
X_D_magnitude = np.abs(X_D_f)

# Step 5: Plot the magnitude spectra and compare

# Plot spectrum of x[n]
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(np.linspace(-0.5, 0.5, len(X_magnitude)), X_magnitude)
plt.title('Magnitude Spectrum of x[n]')
plt.xlabel('Normalized Frequency')
plt.ylabel('|X(f)|')

# Plot spectrum of x_D[n]
plt.subplot(2, 1, 2)
plt.plot(np.linspace(-0.5, 0.5, len(X_D_magnitude)), X_D_magnitude)
plt.title('Magnitude Spectrum of Downsampled x_D[n]')
plt.xlabel('Normalized Frequency')
plt.ylabel('|X_D(f)|')

plt.tight_layout()
plt.show()
```
