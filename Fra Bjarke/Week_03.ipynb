### Problem 8

The zero order hold function can be implemented using a rectangular window of the lenght I. Function created name zoh_interp. 

```python
data = np.array([1,2,3,1,2,3,0])
I = 5
data_up = f.upsample(data, I)
data_zoh = f.zoh_interp(data_up, I)

fig, ax = plt.subplots(figsize=(10,8))
ax.stem(data_up,markerfmt='go',linefmt='g-',label=("Upsampled Data"))
ax.stem(data_zoh,markerfmt='r.', linefmt='r--',label=("Interpolated using ZOH"))
ax.grid()
ax.legend()
ax.set_title("Implementation of Zero-Order-Hold interpolation")
plt.show()
```
